```python

# TASK 1: LOAD AND EXPLORE THE DATASET

import pandas as pd

# Step 1: Loading the dataset (named 'sales_data.csv')
df = pd.read_csv("sales_data.csv")

```

```python
# Step 2: Displays the first few rows
print(df.head())
```

```python
# Step 3: Explore the structure of the dataset
print(df.info())  # Check data types
print(df.isnull().sum())  # Check missing values
```

```python
# Step 4: Clean the dataset
df_cleaned = df.dropna()  # Option 1: Drop missing values
# OR
df_filled = df.fillna(df.mean())  # Option 2: Fill missing values
```

```python
# Display the cleaned dataset
print(df_cleaned.head())  # If dropping
print(df_filled.head())  # If filling
```

```python
#TASK 2: BASIC DATA ANALYSIS

# Step 1: Compute basic statistics
print(df.describe())  # Gives mean, median, std dev, etc.
```

```python
# Step 2: Perform groupings on a categorical column
# Groupings of Month and Profit

grouped_data = df.groupby("Month")["Profit"].mean()
print(grouped_data)
```

```python
# Step 3: Identify patterns or interesting findings
# Checks unique values and their counts to spot trends
print(df["Cost"].value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset 
df = pd.read_csv("sales_data.csv"
```

```python
# Line Chart - Trends over time
import matplotlib.pyplot as plt

# Plotting Month vs. Profit
plt.plot(df['Month'], df['Profit'])
plt.xlabel('Month')
plt.ylabel('Profit')
plt.title('Month vs Profit')
plt.show()
```

```python
# Bar Chart - Comparison across categories

# Plotting a bar chart for Customer Age by Country
df.groupby('Country')['Customer_Age'].mean().plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Customer Age')
plt.title('Customer_Age by Country')
plt.show()
```

```python
# Histogram - Distribution of a numerical column
# Plotting a histogram for Customer_Age

df['Customer_Age'].plot(kind='hist', bins=10)
plt.xlabel('Customer_Age')
plt.title('Customer_Age Distribution')
plt.show()
```

```python

# Plotting a scatter plot of Country vs. Year

df.plot(kind='scatter', x='Country', y='Year')
plt.title('Country vs Year')
plt.show()
```

```python

# Plotting a scatter plot of Year vs. Revenue

df.plot(kind='scatter', x='Year', y='Revenue')
plt.title('Year vs Revenue')
plt.show()
```
